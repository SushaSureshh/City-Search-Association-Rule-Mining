```python
''' This program is authored by Susha Pozhampallan Suresh 

as an assignment for the position of Machine Learning Engineer in Coveo

The program the user data and prints the count of search made by users from a specifi country on hourly basis

I will be using this data to find out the missing country

'''

# import the necessary libraries
import pandas as pd
import datetime

# Read the data
data = pd.read_json('./city_search.json')

# convert the list of session id to data frame 
data['session_id'] = data['session_id'].apply(lambda x: x[0])

# convert a unix timestamp to a naive datetime object
data['unix_timestamp'] = data['unix_timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x[0]))

# find the year, month, day and hour from the time stamp 
data['session_year'] = [datetime.datetime.date(d).year for d in data['unix_timestamp']]
data['session_month'] = [datetime.datetime.date(d).month for d in data['unix_timestamp']]
data['session_day'] = [datetime.datetime.date(d).day for d in data['unix_timestamp']]
data['session_hour'] = [datetime.datetime.time(d).hour for d in data['unix_timestamp']]

# convert the list of user to data frame 
data['user_id'] = data['user'].apply(lambda x: x[0][0]['user_id'])
data['joining_date'] = data['user'].apply(lambda x: x[0][0]['joining_date'])
data['country'] = data['user'].apply(lambda x: x[0][0]['country'])

# drop the user column
data.drop('user', axis=1, inplace=True)

#print the data
data

```

```python
# customize the display

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# display the total number of search performed hourly for all the countries
pd.crosstab(data.country, data.session_hour, margins=True, margins_name="Total")

# Looking at the time zone, it is most likely Russia which is GMT +3
```
